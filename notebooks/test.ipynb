```python
from urllib.parse import urlparse
import ipaddress


def having_ip(url):
    try:
        host = urlparse(url).hostname
        ipaddress.ip_address(host)
        return 1
    except:
        return 0

having_ip('https://aidasdir.github.io/MyEtherWallet/')
```
