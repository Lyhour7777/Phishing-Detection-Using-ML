```python
data = pd.read_csv(r"D:\Backup\I5\I5-ITC\PM\project\Phishing-Detection-Using-ML\data\URL dataset.csv")
data.head()
```

```python
url = data.sample(n = 5000, random_state = 12).copy()
url = url.reset_index(drop=True)
url.head()
```

```python
url.drop(['type'], axis=1, inplace=True)
url.head()
```

```python
features = []
# label = 0

for i in range(0, 20):
  urls = url['url'][i]
  features.append(Preprocess(urls))
```

```python
#converting the list to dataframe
feature_names = ['Have_IP', 'Have_At', 'URL_Length', 'count_subdomain','Redirection', 
                      'https_Domain', 'TinyURL', 'Prefix/Suffix', 'DNS_Record', 
                      'Domain_Age', 'iFrame', 'Mouse_Over','Right_Click', 'Web_Forwards']

extractedData = pd.DataFrame(features, columns= feature_names)
extractedData
```
